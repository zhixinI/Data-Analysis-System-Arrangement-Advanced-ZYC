# day01-matplotlib绘图代码

```python
import numpy as np
import scipy.stats as stats#比numpy多了一个库、数

import matplotlib.style as style
import matplotlib.pyplot as plt
#嵌入，静态图像
%matplotlib inline
#背景版风格设置
style.use('fivethirtyeight')
#字体 清晰度 figure身材，你给长宽尺寸全局设置
plt.rcParams['figure.figsize'] = (14,7)
plt.rcParams['figure.dpi'] = 100
#图形部分
X = np.linspace(-4,4,100)
y = stats.norm.pdf(X) / np.max(stats.norm.pdf(X))

plt.plot(X,y,label='PDF') 
plt.fill_between(X,y,alpha=.3)

plt.plot(X,stats.norm.cdf(X),label='CDF')
#图例#调用
plt.legend() 

#图片上文字的说明
plt.text(x=-1.3,y=.7,s='pdf',rotation=65,alpha=.75,weight='bold',size=20,color='#000FFF')
plt.text(x=.35,y=.5,s='cdf',rotation=55,alpha=.75,weight='bold',size=20,color='#FC4000')

#标题
plt.title('Demo')
#X，Y轴的标签
plt.xlabel('X',color='r')
plt.ylabel('f(x)=exp(-x^2)/(2x)',color='r')
#设置刻度
plt.axis([-4.2,4.2,0,1.2])
#保存图片dpi清晰度
plt.savefig('./pdf.png',dpi=100)
```

```python
'''
labels 标记
shadow 阴影
colors 颜色
explode 脱离
autopct 百分比
'''
import numpy as np
import scipy.stats as stats
import matplotlib.style as style
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#伪动态图像 
%matplotlib inline
#背景版风格设置
style.use('fivethirtyeight')

#字体 清晰度 figure身材，你给长宽尺寸全局设置
plt.rcParams['figure.figsize'] = (8,8)
plt.rcParams['figure.dpi'] = 100
#设置中文字体
#plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['font.family'] = ['Arial Unicode MS']
#中文情况下，负号问题
plt.rcParams['axes.unicode_minus'] =False

X = [50501.17,48500,42000,28000,24000,20000,19000,17000,16000,15000]
lables = ['广东','江苏','山东','浙江','河南','四川','湖北','湖南','河北','福建']

#找到GDP最高的省份
explode = np.zeros(shape=len(X))
max_index = np.argmax(X)
#稍微脱离整个半径的0.1倍
explode[max_index] = .1

#左右两个%，标签
a = plt.pie(X,labels=lables,autopct='%.2f%%',explode=explode,colors=sns.color_palette("autumn",n_colors=len(X)))
```

#### 雷达图 

```python
import numpy as np
import scipy.stats as stats#比numpy多了一个库、数
import matplotlib.style as style
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
#伪动态图像
%matplotlib inline
#背景版风格设置
style.use('fivethirtyeight')

#字体 清晰度 figure身材，你给长宽尺寸全局设置
plt.rcParams['figure.figsize'] = (8,8)
plt.rcParams['figure.dpi'] = 100

#设置中文字体
#plt.rcParams['font.family'] = ['SimHei']
plt.rcParams["font.family"] = 'Arial Unicode MS'
#中文情况下，负号问题
plt.rcParams['axes.unicode_minus'] = False

#标签
labels = np.array(['推塔','生存','承受伤害','经济','团战','输出'])
#数据个数6个labels数据得对应
dataLen = labels.size

#数据 李白 
data = np.array([1,6,3,7,4,8])
#合并时维度需一样，
data=np.concatenate((data,[data[0]]))

#角度2*np.pi是圆。
angles = np.linspace(0,2*np.pi,dataLen,endpoint=False)
angles = np.concatenate((angles,[angles[0]]))
#画布尺寸对象
fig = plt.figure()

#定义极图,子图位置  111  121  211  221 122偏移位置polar=True告诉你我要画了
ax = fig.add_subplot(111,polar=True)
#雷达的边界线
ax.plot(angles,data,'co-',lw=2)
#内部填充
ax.fill(angles,data,facecolor='c',alpha=.2)

#设置6个参数
#ax.set_thetagrids(angles * 180 / np.pi,labels,fontproperties='SimHei',size=20)
ax.set_thetagrids(angles * 180 / np.pi,labels,fontproperties='Arial Unicode MS',size=20)

ax.set_title('李白')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#k线图时间单位必须整数
import matplotlib.dates as dates
import mpl_finance as mf

plt.rcParams['figure.figsize'] = [18,6]
#设置中文
sns.set_style('darkgrid',{'font.sans-serif':['Arial Unicode MS','Arial']})

df_stock=pd.read_excel('../data/茅台股份.xlsx')
#将时间作为行索引
df_stock.set_index('日期',inplace=True)

#OHLC限制，成交量、换手率等不画在图中
new_df = df_stock.loc[:,['开盘','最高','最低','收盘']]
new_df = new_df.iloc[:100]

#转换时间为整数
#默认的dr series ndarray  list tuple 只认识zip() 生成器(整数时间，O,H,L,C)
zip_data = zip(dates.date2num(new_df.index.to_pydatetime()),new_df.开盘,new_df.最高,new_df.最低,new_df.收盘)

#k线图和3D图有相同的特质，都没有尺寸对象
ax = plt.gca()
plt.title('茅台股份OHLC-K线图')
mf.candlestick_ohlc(ax,zip_data,colorup='r',colordown='g',width=5)

#必须把时间调整回来
ax.xaxis_date()

#把X轴倾斜
plt.xticks(rotation=45)
```
